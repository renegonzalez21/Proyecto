### 2.3 Modelo que representa el problema

#### Situación en Guadalajara
Consideremos la rehabilitación de diferentes avenidas en Guadalajara, Jalisco, que presenta un alto volumen de tráfico y deterioro en el pavimento existente. La administración de la ciudad ha decidido evaluar diferentes diseños de pavimentos para optimizar costos y asegurar una mayor durabilidad en condiciones específicas de tráfico y clima.

#### Modelo Matemático

**Función a Optimizar**:
$C = C_m + C_c + C_{mo}$

Donde:

$ C $ :es el costo total del pavimento.

$C_m$ :es el costo de los materiales.

$C_c$ :es el costo de la construcción.

$C_{mo} $ :es el costo de mantenimiento a lo largo de la vida útil.

**Restricciones**:
1. **Condiciones de carga**: La resistencia del pavimento debe ser suficiente para soportar cargas de tráfico. Esto se puede expresar como:
   $ R_p \geq T$


   Donde:


   $R_p$ :es la resistencia del pavimento.


   $T$: es la carga de tráfico esperada (determinado a partir de estudios de tráfico).

3. **Durabilidad**: La vida útil del pavimento debe ser al menos de 10 años:


   $V \geq 10$


   Donde:


   $V$: es la vida útil proyectada del pavimento.

6. **Espesor mínimo y máximo**: El espesor $e$ del pavimento debe cumplir con ciertas especificaciones:


   $e_{min} \leq e \leq e_{max}$

#### Deducción del Modelo
1. **Costos de Materiales**: Se define un costo por unidad de material $c_m$ y se relaciona con el espesor $e$:


   $C_m = c_m \cdot A \cdot e$


   Donde $A$ es el área del pavimento.

3. **Costos de Construcción**: Se estima un costo por unidad de trabajo $c_c$ para la construcción:


   $C_c = c_c \cdot A $

5. **Costos de Mantenimiento**: Se puede estimar un costo anual de mantenimiento $c_{mo}$:


   $C_{mo} = c_{mo} \cdot V$

Combinando estas ecuaciones, la función de costo total se puede expresar como:

$C = (c_m \cdot A \cdot e) + (c_c \cdot A) + (c_{mo} \cdot V)$

#### Situación que Representa el Modelo
El modelo representa la decisión de diseñar un pavimento que sea económicamente viable y duradero para una avenida en Guadalajara con alta carga de tráfico. Busca equilibrar los costos iniciales y los costos a lo largo de la vida útil del pavimento.

#### Limitaciones Fundamentales
- **Variabilidad de materiales**: La disponibilidad y el costo de los materiales pueden variar según el mercado.
- **Condiciones ambientales**: Factores como clima, humedad y temperatura pueden afectar la durabilidad real del pavimento.
- **Cargas de tráfico**: Cambios inesperados en el volumen y tipo de tráfico durante la vida útil pueden influir en el rendimiento.

#### Significado y Valor de los Parámetros
- $c_m$: Costo por unidad de material (dólares por metro cúbico).
- $c_c$: Costo por unidad de construcción (dólares por metro cuadrado).
- $c_mo$: Costo de mantenimiento anual (dólares por año).
- $A$: Área total a pavimentar (metros cuadrados).
- $e$: Espesor del pavimento (metros).
- $R_p$: Resistencia del pavimento (kilonewtons).
- $T$: Carga de tráfico esperada (kilonewtons).
- $ V$: Vida útil proyectada (años).



### 2.4 Solución del problema de optimización

### Situación Real: Rehabilitación de la Avenida López Mateos

**Contexto**:
La Avenida López Mateos es una de las arterias principales de Guadalajara, conocida por su alto volumen de tráfico y su importancia como vía de conexión entre diversas zonas de la ciudad. Debido al tráfico intenso y a las condiciones climáticas de la región, el pavimento ha mostrado signos de deterioro, lo que hace necesaria su rehabilitación.

**Datos Reales Recopilados**:

1. **Área de Pavimentación (A)**:
   - Aproximadamente 12,000 m², considerando un tramo específico que requiere rehabilitación.

2. **Carga de tráfico esperada (T)**:
   - 35,000 vehículos diarios, con una carga promedio estimada en 85 kN por vehículo.

3. **Costos**:
   - **Costo de materiales ($c_m$)**: 35 por m³ (promedio de costos de materiales usados en pavimentos de la región).
   - **Costo de construcción ($c_c$)**: 30 por m².
   - **Costo de mantenimiento anual ($c_{mo}$)**: 2500 por año.

4. **Durabilidad mínima (V)**:
   - La vida útil esperada del pavimento debe ser al menos de 15 años, dado el alto tráfico.

5. **Especificaciones de espesor**:
   - **Espesor mínimo ($e_{min}$)**: 0.15 m.
   - **Espesor máximo ($e_{max}$)**: 0.25 m.

6. **Resistencia del pavimento (R_p)**:
   - Para esta carga de tráfico, se requiere una resistencia mínima de 90 kN.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import linprog

A = 12000 
T = 35000 
c_m = 35   
c_c = 30   
c_mo = 2500 
V_min = 15 
e_min = 0.15  
e_max = 0.25 
R_p_min = 90

c = [c_m * A + c_c * A + np.mean(c_mo), np.mean(c_mo) * V_min] 

A_ub = [
    [-1, 0],  
    [1, 0],   
    [0, -1], 
]
b_ub = [
    -e_min,   
    e_max,    
    -V_min,   
]


x0_bounds = (e_min, e_max) 
x1_bounds = (V_min, None)

x0_bounds = (e_min, e_max) 
x1_bounds = (V_min, None)

resultado_Lopez_Mateos = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=[x0_bounds, x1_bounds], method='highs')

resultado_Lopez_Mateos
```

```python
espesor_optimo = resultado_Lopez_Mateos.x[0]
vida_util = resultado_Lopez_Mateos.x[1]
costo_total = resultado_Lopez_Mateos.fun
    
espesor_optimo, vida_util, costo_total
```

* __Espesor óptimo del pavimento__: 0.15 m
* __Vida útil proyectada__: 15.0 años
* __Costo total estimado__: $679,875.00

### Interpretación de los Resultados
- Espesor óptimo (0.15 m):

El espesor de 0.15 m es el mínimo permitido según las especificaciones establecidas. Esto indica que, para cumplir con las restricciones de resistencia y durabilidad, no es necesario aumentar el espesor del pavimento. Sin embargo, al estar en el límite inferior, se debe considerar que un espesor mayor podría ofrecer mejores características de durabilidad y resistencia, especialmente dado el alto tráfico en la Avenida López Mateos.

- Vida útil proyectada (15.0 años):

La vida útil de 15 años, esto nos sugiere que, bajo las condiciones actuales y considerando un mantenimiento adecuado, el pavimento debería durar al menos este tiempo antes de requerir una rehabilitación significativa.

- Costo total estimado ($679,875.00):

El costo total de $679,875.00 es significativo, pero no inusualmente alto para un proyecto de rehabilitación de pavimentos en una avenida principal. Este costo incluye materiales, construcción y costos de mantenimiento a lo largo de la vida útil proyectada.
